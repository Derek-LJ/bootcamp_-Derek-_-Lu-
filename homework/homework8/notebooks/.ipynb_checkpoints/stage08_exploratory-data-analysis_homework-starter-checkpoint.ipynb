```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import skew, kurtosis
sns.set(context='talk', style='whitegrid')
np.random.seed(8)
pd.set_option('display.max_columns', 100)

# === Synthetic data generator (adapt or replace with your own data) ===
n = 160
df = pd.DataFrame({
    'date': pd.date_range('2021-02-01', periods=n, freq='D'),
    'region': np.random.choice(['North','South','East','West'], size=n),
    'age': np.random.normal(40, 8, size=n).clip(22, 70).round(1),
    'income': np.random.lognormal(mean=10.6, sigma=0.3, size=n).round(2),
    'transactions': np.random.poisson(lam=3, size=n),
})
base = df['income'] * 0.0015 + df['transactions']*18 + np.random.normal(0, 40, size=n)
df['spend'] = np.maximum(0, base).round(2)

# inject a bit of missingness and outliers
df.loc[np.random.choice(df.index, 5, replace=False), 'income'] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), 'spend'] = np.nan
df.loc[np.random.choice(df.index, 2, replace=False), 'transactions'] = df['transactions'].max()+12
df.head()
```

```python
desc = df[['age','income','transactions','spend']].describe().T
desc['skew'] = [skew(df[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(df[c].dropna()) for c in desc.index]
desc
```

## 6) Insights & Assumptions (write your commentary)
### Insights & Assumptions

**Top 3 Insights**
1. **Income distribution** is roughly bell-shaped but right-skewed, with most customers clustered between 30kâ€“50k and a few high-income outliers up to 87k.  
2. **Transactions** are low on average (mean â‰ˆ 3), but a few outliers exist with >15 transactions, which may represent unusual customer behavior or data entry errors.  
3. **Spending** is moderately correlated with both **income (0.31)** and **transactions (0.48)**. This suggests that both financial capacity and activity frequency contribute to higher spend, though the relationship is not strictly linear.  

**Assumptions & Risks**
- We assume missing values (e.g., income and spend) are **missing at random**, and can be imputed without introducing systematic bias. If this assumption is wrong, imputations may distort relationships.  
- We assume numeric variables (income, spend, transactions) are measured consistently across regions. If regional reporting differs, observed correlations may be biased.  
- Outliers in transactions and spend may represent **true high-value customers**, not errors. Aggressively removing them risks discarding important business cases.  

**Next Steps Before Modeling**
- **Data Cleaning**: Impute missing values (median for skewed numeric data, mode for categorical).  
- **Feature Engineering**: Create new features such as `income_per_transaction` or `spend_to_income_ratio`.  
- **Scaling**: Normalize or standardize skewed variables (e.g., income, spend) to improve model performance.  
- **Categorical Encoding**: Convert `region` into dummy variables or use target encoding for modeling.  
- **Outlier Strategy**: Decide between winsorization, capping, or retaining outliers depending on model robustness and business needs.
